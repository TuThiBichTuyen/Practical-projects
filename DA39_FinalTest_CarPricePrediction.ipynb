# Problem Statement:
Một công ty ô tô Trung Quốc Geely Auto muốn thâm nhập thị trường Mỹ bằng cách thành lập đơn vị sản xuất của họ ở đó và sản xuất ô tô để cạnh tranh với các đối tác Mỹ và châu Âu.
Do vậy, họ đã ký hợp đồng với một công ty tư vấn ô tô để hiểu các yếu tố ảnh hưởng đến việc định giá ô tô tại thị trường Mỹ, vì những yếu tố đó có thể rất khác so với thị trường Trung Quốc. Cụ thể, họ muốn biết:
➔ Đâu là những yếu tố quan trọng để định giá ô tô?
➔ Các thông số nào của ô tô sẽ ảnh hưởng đến giá của chiếc ô tô đó?
# Business Goal:
Bạn được yêu cầu xây dựng một mô hình giá ô tô với các features trong tập dữ liệu. Nó sẽ được team Product sử dụng để hiểu chính xác mức giá thay đổi như thế nào với các features đó. Theo đó, họ có thể đưa ra các phương án thiết kế của ô tô, chiến lược kinh doanh, v.v. để đáp ứng
với từng phân khúc giá nhất định. Hơn nữa, mô hình này sẽ là một cách tốt để BOD hiểu được cách định giá của một thị trường mới.

# Giải thích thuật ngữ:
+ Symboling: Đánh giá rủi ro bảo hiểm, giá trị +3 cho thấy xe ô tô có rủi ro cao, trong khi -3 cho thấy nó có khả năng an toàn khá cao.
+ Fueltype: loại nhiên liệu
  + gas
  + diesel
+ Aspiration: hệ thống hút không khí vào động cơ
  + Std: Đây là hệ thống hút không khí thông thường, trong đó không có sự can thiệp từ bất kỳ thiết bị nào như turbocharger hay supercharger. Động cơ hút tự nhiên phụ thuộc vào sự hút không khí tự nhiên để hoạt động.
  + Turbo: Đây là hệ thống hút mạnh mẽ hơn, thường sử dụng turbocharger hoặc supercharger để tăng áp suất không khí đưa vào động cơ. Bằng cách nạp thêm khí nén, hút ép giúp tăng cường hiệu suất động cơ và cung cấp công suất tốt hơn.
+ Doornumber: số cửa
  + Two: Một số xe thể thao hoặc xe hơi đặc biệt có thiết kế chỉ với 2 cửa, một cửa đối với hành khách phía trước và một cửa đối với hành khách phía sau.
  + Four: Đây là một lựa chọn phổ biến cho các loại xe hơi thông thường. Có cả cửa phía trước và cửa phía sau cho hành khách.
+ Carbody: kiểu dáng và cấu trúc bên ngoài của một chiếc xe, gồm có 5 loại sau
  + sedan: Xe sedan có kiểu dáng thân xe 3 khoang hoặc 4 khoang với một khoang hành lý phía sau. Nó có một khoang ngồi riêng biệt cho hàng ghế sau và cửa sau độc lập. Sedan thường có thiết kế sang trọng và phù hợp cho việc đi lại hàng ngày hoặc sử dụng cá nhân.
  + hatchback: Xe hatchback có một khoang hành lý tích hợp với phần thân xe, và cửa hành lý mở bằng cách nâng cả cửa sau. Nó có thiết kế linh hoạt và thường có không gian nội thất rộng hơn so với sedan cùng phân khúc.    
  + wagon (xe du lịch) là một kiểu dáng thân xe có khoang hành lý dài và phẳng, mở rộng từ phần sau của xe và tích hợp với phần thân xe. Xe wagon thường có kiểu dáng giống với sedan ở phần trước, nhưng có một khoang hành lý lớn và dễ truy cập ở phía sau. Wagon thường có không gian nội thất rộng rãi và khả năng chở hàng hóa lớn, làm cho nó trở thành lựa chọn phổ biến cho gia đình hoặc những người có nhu cầu vận chuyển nhiều đồ đạc hoặc hàng hóa. Một số wagon cũng có khả năng gập hàng ghế sau để tăng không gian lưu trữ.          
  + hardtop (mui cứng): là một kiểu mui xe trong đó không có khung chắn giữa các cửa sổ bên. Thay vì có khung cửa sổ bên nổi bật, mui cứng bao gồm một mảng cứng liền mạch từ mui trước đến mui sau của xe. Mui cứng thường được làm bằng vật liệu cứng như thép hoặc nhôm và có thể không thể gập lại hoặc có thể gập lại tự động. Khi mui cứng không thể gập lại, nó tạo ra một diện mạo mượt mà và thường liên kết với các loại xe thể thao hoặc xe hơi cao cấp. Mui cứng có thể mang lại sự chắc chắn và cảm giác an toàn hơn so với mui mềm.
  + convertible (xe mui trần): Convertible có khả năng mở rộng hoặc gập được mui xe, cho phép cảm giác lái trong không gian ngoài trời. Có thể có nhiều kiểu mui trần khác nhau, từ mui cứng có thể gập lại tự động đến mui mềm có thể tháo rời.
+ Drivewheel: hệ thống dẫn động của xe
  + fwd (Front-wheel drive): Trong hệ thống này, động cơ truyền lực đến bánh xe trước. Bánh xe sau không có nguồn lực riêng để di chuyển xe. Đây là hệ thống phổ biến trong nhiều loại xe hơi, đặc biệt là xe hạng nhỏ và xe gia đình.
  + rwd (Rear-wheel drive): Trong hệ thống này, động cơ truyền lực đến bánh xe sau. Bánh xe trước chỉ có nhiệm vụ lái xe. Hệ thống dẫn động cầu sau thường được sử dụng trong các xe thể thao và xe hạng sang, với trọng lượng phân bố lý tưởng và khả năng tăng tốc tốt.
  + 4wd (Four-wheel drive): Trong hệ thống này, động cơ có thể truyền lực đến cả bánh xe trước và bánh xe sau. Hệ thống này cung cấp khả năng tăng cường độ bám đường và vận hành trên đa dạng điều kiện đường. Nó thường được sử dụng trong các xe địa hình, xe thể thao đa dụng và xe hơi cao cấp.
+ Enginelocation : vị trí đặt động cơ. Có hai vị trí chính cho động cơ trong một chiếc xe hơi:
  + front (động cơ phía trước): Trong trường hợp này, động cơ được đặt ở phía trước xe, thường là phía trước trục lái. Đây là vị trí phổ biến nhất trong hầu hết các loại xe hơi, bao gồm xe hạng nhỏ, xe gia đình, xe thể thao và nhiều loại khác.
  + rear (động cơ phía sau): Trong trường hợp này, động cơ được đặt ở phía sau xe, thường là phía sau trục lái. Đây là một vị trí đặc biệt và thường được sử dụng trong một số loại xe đặc biệt như một số mẫu xe thể thao Porsche 911 và xe hơi cổ Volkswagen Beetle.
+ Wheelbase: (khoảng cách giữa hai trục bánh xe) là khoảng cách từ trục bánh xe trước đến trục bánh xe sau trên một chiếc xe. Nó được đo từ trung tâm trục bánh xe trước đến trung tâm trục bánh xe sau.Khoảng cách giữa hai trục bánh xe quyết định phần lớn về tổng chiều dài của một chiếc xe và có ảnh hưởng đến nhiều yếu tố khác nhau, bao gồm:
  + Khả năng quay vòng: Khoảng cách giữa hai trục bánh xe ảnh hưởng đến bán kính quay vòng của xe. Một khoảng cách giữa hai trục bánh xe nhỏ hơn giúp xe xoay quanh một đường cung nhỏ hơn và có khả năng quay vòng nhỏ hơn.
  + Ổn định và cảm giác lái: Khoảng cách giữa hai trục bánh xe cũng ảnh hưởng đến ổn định và cảm giác lái của xe. Một khoảng cách lớn hơn giữa hai trục bánh xe thường mang lại sự ổn định tốt hơn ở tốc độ cao, trong khi một khoảng cách nhỏ hơn giữa hai trục bánh xe thường mang lại cảm giác lái thể thao hơn.
  + Không gian nội thất: Khoảng cách giữa hai trục bánh xe cũng có ảnh hưởng đến không gian nội thất của xe, đặc biệt là không gian chân cho hành khách ở hàng ghế sau.
+ Carlength: chiều dài xe
+ Carwidth: chiều rộng xe
+ Carheight: chiều cao xe
+ Curbweight (trọng lượng tự nhiên): là trọng lượng của một chiếc xe khi nó không có hành khách, hàng hoặc tải trọng khác
+ Enginetype: loại động cơ
  + ohc (Overhead Camshaft:Trục cam trên đầu): Trong động cơ OHC, trục cam nằm trên đầu xi-lanh và điều khiển các van đầu. Đây là một thiết kế phổ biến trong các động cơ hiện đại, nó giúp cải thiện hiệu suất và tăng cường khả năng vận hành so với các thiết kế trục cam nằm dưới.
  + ohcf (Overhead Camshaft with Fuel Injection): OHCF kết hợp cả trục cam trên đầu (OHC) và hệ thống phun nhiên liệu trực tiếp (Fuel Injection). Hệ thống phun nhiên liệu trực tiếp giúp tăng cường hiệu suất và hiệu quả nhiên liệu bằng cách cung cấp nhiên liệu trực tiếp vào xi-lanh thay vì thông qua hệ thống phun nhiên liệu thông thường.
  + ohcv (Overhead Camshaft with Variable Valve Timing): OHCV là một biến thể của động cơ OHC nhưng có thêm công nghệ điều chỉnh thời gian mở/đóng van (Variable Valve Timing). Công nghệ này cho phép điều chỉnh thời gian mở/đóng van để tối ưu hóa hiệu suất động cơ ở các dải tốc độ khác nhau, cung cấp cả hiệu suất cao và tiết kiệm nhiên liệu.
  + dohc (Dual Overhead Camshaft): DOHC là một thiết kế động cơ có hai trục cam trên đầu. Mỗi trục cam điều khiển một nhóm van, một để điều khiển van hút và một để điều khiển van xả. DOHC thường được sử dụng trong các động cơ hiệu suất cao, cho phép việc kiểm soát chính xác hơn của van và tăng khả năng vận hành của động cơ.
  + l:  Động cơ "L" là một thiết kế động cơ có hình dạng giống chữ "L" khi nhìn từ phía trên. Nó thường được sử dụng trong các động cơ xe máy và xe máy phân khối lớn. Thiết kế "L" giúp tối ưu hóa không gian và cân bằng trọng lượng của động cơ.
  + rotor: (Rotary Engine): Động cơ rotor, hay còn được gọi là động cơ Wankel, là một loại động cơ không có xi-lanh cổ điển. Thay vào đó, nó sử dụng một bộ phận quay gọi là rotor để tạo ra chuyển động và công suất. Động cơ rotor thường nhỏ gọn, nhẹ và có kích thước nhỏ, và thường được sử dụng trong các ứng dụng đặc biệt như các mẫu xe Mazda RX-7 và RX-8.
  + dohcv (Dual Overhead Camshaft with Variable Valve Timing): DOHCV là một biến thể của động cơ DOHC kết hợp với công nghệ điều chỉnh thời gian mở/đóng van (Variable Valve Timing). Điều này cho phép việc điều chỉnh thời gian mở/đóng van để tối ưu hóa hiệu suất và tiết kiệm nhiên liệu trong các động cơ có hai trục cam trên đầu.
+ Cylindernumber: số lượng xi-lanh có trong động cơ của một chiếc xe. Xi-lanh là một phần quan trọng của động cơ và giúp tạo ra công suất bằng cách chứa hỗn hợp nhiên liệu-khí và cháy để tạo ra sự chuyển động. Số xi-lanh càng lớn, thường thể hiện mức công suất và hiệu suất cao hơn của động cơ. Số lượng xi-lanh ảnh hưởng đến khả năng sản sinh công suất và mô-men xoắn của động cơ. Động cơ nhiều xi-lanh thường có khả năng sản sinh công suất lớn hơn và hoạt động mượt mà hơn so với động cơ ít xi-lanh. Số xi-lanh cũng ảnh hưởng đến kích thước, trọng lượng và kích thước tổng thể của động cơ. Động cơ có số xi-lanh ít hơn thường nhỏ gọn hơn và nhẹ hơn, trong khi động cơ có số xi-lanh nhiều hơn thường có kích thước và trọng lượng lớn hơn. Số xi-lanh cũng có thể ảnh hưởng đến tiêu thụ nhiên liệu và lượng khí thải của động cơ. Động cơ có số xi-lanh ít hơn thường có xu hướng tiêu thụ ít nhiên liệu hơn, trong khi động cơ có số xi-lanh nhiều hơn thường tiêu thụ nhiên liệu nhiều hơn, đặc biệt khi hoạt động ở tốc độ cao.

+ Enginesize (kích thước động cơ): thường được đo bằng dung tích xi-lanh của động cơ. Nó thể hiện tổng dung tích của tất cả các xi-lanh trong động cơ và được đo bằng đơn vị đo lường thể tích như đặc biệt, centimet khối (cc), hoặc lít (L).
Kích thước động cơ là một yếu tố quan trọng ảnh hưởng đến công suất và hiệu suất của động cơ. Thông thường, động cơ có kích thước lớn hơn có xu hướng sản sinh công suất và mô-men xoắn lớn hơn. Tuy nhiên, nó cũng có thể liên quan đến tiêu thụ nhiên liệu cao hơn.
Kích thước động cơ có thể ảnh hưởng đến khả năng tăng tốc và tốc độ tối đa của một chiếc xe. Một động cơ lớn hơn thường cung cấp sự tăng tốc mạnh mẽ và đạt được tốc độ cao hơn, trong khi động cơ nhỏ hơn thường phù hợp cho việc vận hành tiết kiệm nhiên liệu và di chuyển trong thành phố.
+ Fuelsystem: Hệ thống nhiên liệu trong một xe ô tô là hệ thống được thiết kế để cung cấp nhiên liệu từ bình chứa nhiên liệu đến động cơ để đáp ứng nhu cầu vận hành. Goồm cácc loại sau:
  + MPFI (Multi-Point Fuel Injection): MPFI là hệ thống phun nhiên liệu đa điểm. Trong hệ thống này, mỗi xi-lanh được trang bị một injector riêng biệt để phun nhiên liệu trực tiếp vào đường vào của xi-lanh. MPFI cung cấp kiểm soát chính xác hơn về lượng nhiên liệu và tạo điều kiện cho đốt cháy hiệu quả hơn, cải thiện hiệu suất và tiêu thụ nhiên liệu.
  + 2BBL (Two-Barrel Carburetor): 2BBL là một loại bộ chế hòa khí có hai cọc (barrel). Bộ chế hòa khí 2BBL cung cấp nhiên liệu và không khí cho động cơ thông qua hai cọc riêng biệt. Mỗi cọc có một bộ cảm biến và van điều khiển nhiên liệu riêng, giúp cung cấp nhiên liệu và không khí một cách cân bằng.
  + IDI (Indirect Injection): IDI là hệ thống phun nhiên liệu gián tiếp. Trong hệ thống này, nhiên liệu được phun vào một không gian phụ bên trong buồng đốt, từ đó nó được hòa trộn với không khí và cháy. Hệ thống IDI thường được sử dụng trong động cơ diesel.
  + 1BBL (One-Barrel Carburetor): 1BBL là một loại bộ chế hòa khí có một cọc. Bộ chế hòa khí 1BBL cung cấp nhiên liệu và không khí cho động cơ thông qua một cọc duy nhất. Mặc dù ít phổ biến hơn so với các hệ thống khác, nhưng nó vẫn được sử dụng trong một số ứng dụng đơn giản và xe cổ điển.
  + SPDI (Sequential Port Direct Injection): SPDI là hệ thống phun nhiên liệu trực tiếp tuần tự. Trong hệ thống này, nhiên liệu được phun trực tiếp vào các van hút của mỗi xi-lanh theo thứ tự tuần tự. SPDI cung cấp kiểm soát chính xác hơn về lượng nhiên liệu và giúp tăng cường hiệu suất và hiệu quả đốt cháy.
  + 4BBL (Four-Barrel Carburetor): 4BBL là một loại bộ chế hòa khí có bốn cọc. Bộ chế hòa khí 4BBL cung cấp nhiên liệu và không khí cho động cơ thông qua bốn cọc riêng biệt. Hệ thống này cho phép cung cấp lượng nhiên liệu lớn hơn và điều khiển tốt hơn cho các động cơ có nhu cầu công suất cao.
  + MFI (Multi-Point Fuel Injection): MFI là một hệ thống phun nhiên liệu đa điểm tương tự như MPFI. Hệ thống MFI cung cấp nhiên liệu trực tiếp vào mỗi xi-lanh thông qua các injector riêng biệt. Điều này cung cấp sự kiểm soát chính xác hơn về lượng nhiên liệu và cải thiện hiệu suất và tiêu thụ nhiên liệu.
  + SPFI (Sequential Port Fuel Injection): SPFI là một hệ thống phun nhiên liệu trực tiếp tuần tự tương tự như SPDI. Nhiên liệu được phun trực tiếp vào các van hút của từng xi-lanh theo thứ tự tuần tự, cung cấp kiểm soát chính xác hơn về lượng nhiên liệu và tăng cường hiệu suất đốt cháy.
+ Boreratio (tỉ số đường kính xylanh): mô tả sự tương quan giữa đường kính của xi-lanh và hành trình của piston trong một động cơ. Nó được tính bằng cách chia đường kính của xi-lanh cho hành trình piston.
Tỉ số đường kính xylanh (bore ratio) có thể ảnh hưởng đến hiệu suất, công suất và đặc tính vận hành của động cơ. Các tỉ số khác nhau có thể phù hợp với các ứng dụng và yêu cầu khác nhau.
Nếu tỉ số đường kính xylanh cao, tức là đường kính của xi-lanh lớn hơn hành trình piston, thì động cơ có xu hướng làm việc tốt ở tốc độ cao, tạo điều kiện cho lưu lượng khí nạp và đốt cháy nhanh. Điều này có thể dẫn đến công suất cao hơn và tăng hiệu suất động cơ ở tốc độ cao. Tuy nhiên, tỉ số đường kính xylanh cao cũng có thể làm giảm mô-men xoắn ở tốc độ thấp và tạo ra ít lực đẩy ở dải vòng tua thấp.
Ngược lại, nếu tỉ số đường kính xylanh thấp, tức là đường kính của xi-lanh nhỏ hơn hành trình piston, thì động cơ có xu hướng làm việc tốt ở tốc độ thấp và tạo mô-men xoắn mạnh ở dải vòng tua thấp. Điều này phù hợp cho các ứng dụng yêu cầu lực đẩy mạnh ở tốc độ thấp, ví dụ như xe tải hoặc xe địa hình.
+ Stroke (hành trình piston) là khoảng cách mà piston di chuyển trong xi-lanh của một động cơ đốt trong, từ vị trí trên cùng (đỉnh) đến vị trí dưới cùng (đáy) và ngược lại. Nó là một trong những yếu tố quan trọng trong thiết kế động cơ và ảnh hưởng đến hiệu suất và đặc tính vận hành của nó. Hành trình piston ảnh hưởng trực tiếp đến dung tích xi-lanh và tỉ số nén của động cơ. Dung tích xi-lanh được tính bằng cách nhân diện tích của xi-lanh với hành trình piston. Tỉ số nén là tỉ lệ giữa dung tích xi-lanh lúc nén và dung tích xi-lanh lúc xả. Hành trình piston càng lớn thì dung tích xi-lanh cũng càng lớn, điều này có thể tạo ra một số lợi ích như:
  + Tăng công suất: Dung tích xi-lanh lớn hơn cho phép hỗn hợp nhiên liệu-khí nạp lớn hơn và tạo ra sức ép cao hơn trong quá trình cháy. Điều này có thể dẫn đến tăng công suất của động cơ.
  + Tăng mô-men xoắn: Hành trình piston dài có thể tạo ra một cung đường dài hơn cho đốt cháy, cung cấp mô-men xoắn mạnh mà động cơ có thể tạo ra ở tốc độ thấp.
  + Tiêu thụ nhiên liệu tốt hơn: Với dung tích xi-lanh lớn, động cơ có thể hoạt động ở tốc độ thấp hơn để đạt được cùng một công suất. Điều này có thể giúp giảm tiêu thụ nhiên liệu.
Tuy nhiên, hành trình piston càng dài cũng có thể gây ra một số hạn chế như:
  + Tốc độ động cơ giới hạn: Hành trình piston dài có thể giới hạn tốc độ tối đa mà động cơ có thể hoạt động do quá trình di chuyển của piston.
  + Động cơ nặng hơn: Hành trình piston dài yêu cầu thiết kế xi-lanh và bộ phận liên quan lớn hơn, dẫn đến trọng lượng tổng thể của động cơ lớn hơn.
+ Compressionratio (tỉ số nén) là một đại lượng trong động cơ đốt trong, nó biểu thị tỉ lệ giữa dung tích xi-lanh lúc nén và dung tích xi-lanh lúc xả. Tỉ số nén có ảnh hưởng lớn đến hiệu suất và đặc tính vận hành của động cơ.
Tỉ số nén được tính bằng cách chia dung tích xi-lanh lúc nén cho dung tích xi-lanh lúc xả. Dung tích xi-lanh lúc nén là dung tích của không gian trong xi-lanh khi piston đạt đến vị trí trên cùng (đỉnh) sau quá trình nén. Dung tích xi-lanh lúc xả là dung tích khi piston ở vị trí dưới cùng (đáy) sau quá trình xả.
Tỉ số nén có ảnh hưởng đến hiệu suất và đặc tính vận hành của động cơ như sau:
  + Hiệu suất: Tỉ số nén cao có thể cải thiện hiệu suất của động cơ. Khi hỗn hợp nhiên liệu-khí nạp bị nén, áp suất và nhiệt độ tăng lên. Điều này làm tăng hiệu suất cháy và hiệu suất nhiệt của động cơ.
  + Công suất: Tỉ số nén cao có thể cung cấp công suất cao hơn cho động cơ. Khi áp suất cháy tăng, lực đẩy trên piston tăng, dẫn đến công suất động cơ tăng lên.
  + Tiêu thụ nhiên liệu: Tỉ số nén cao có thể giảm tiêu thụ nhiên liệu. Khi áp suất và nhiệt độ tăng lên, hỗn hợp nhiên liệu-khí nạp cháy tốt hơn và cung cấp năng lượng cao hơn từ một lượng nhiên liệu nhất định.
  + Khả năng chống kẹt: Tỉ số nén cao có thể cải thiện khả năng chống kẹt của động cơ. Khi hỗn hợp nhiên liệu-khí nạp bị nén chặt, khả năng phản xạ ngược của ngọn lửa được giảm, giúp ngăn chặn hiện tượng kẹt cháy.
Tuy nhiên, tỉ số nén cũng có một số hạn chế:
  + Nhiệt độ cao: Tỉ số nén cao gây ra áp suất và nhiệt độ cao trong xi-lanh, có thể gây ra vấn đề về nhiệt độ và làm tăng nguy cơ kẹt cháy ngọn lửa tự động (động kinh).
  + Yêu cầu nhiên liệu cao cấp: Tỉ số nén cao có thể yêu cầu sử dụng nhiên liệu cao cấp hơn để đảm bảo khả năng chống kẹt và hiệu suất tối đa.
+ Horsepower: công suất động cơ xe
+ Peakrpm (revolutions per minute)(hiệu suất): là giá trị tốc độ quay tối đa mà một động cơ có thể hoạt động một cách hiệu quả và đạt công suất tối đa. Nó đại diện cho điểm mà động cơ đạt đến số vòng quay tối đa trước khi gặp hạn chế hoặc giảm hiệu suất. Giá trị Peak RPM là riêng cho từng động cơ và thường được nhà sản xuất cung cấp làm thông số hiệu suất. Nó cho biết phạm vi hoạt động tối ưu của động cơ để đạt công suất và mô-men xoắn cực đại. Hoạt động động cơ trong hoặc gần phạm vi Peak RPM có thể cải thiện khả năng tăng tốc và hiệu suất tổng thể.
Giá trị Peak RPM có thể thay đổi tùy thuộc vào thiết kế và mục đích của động cơ. Ví dụ, trong các xe thể thao có hiệu suất cao, Peak RPM thường cao hơn, cho phép động cơ tạo ra công suất lớn hơn ở tốc độ cao. Ngược lại, các động cơ được thiết kế cho các ứng dụng công việc nặng như xe tải hoặc máy móc công nghiệp có thể có giá trị Peak RPM thấp hơn nhưng tập trung vào mô-men xoắn.
Cần lưu ý rằng hoạt động động cơ ở hoặc gần giá trị Peak RPM trong thời gian dài có thể làm tăng mức độ hao mòn và tiêu thụ nhiên liệu, và có thể làm giảm tuổi thọ động cơ. Ngoài ra, hiệu suất nhiên liệu có thể giảm ở RPM cao do tiêu thụ nhiên liệu tăng lên.
+ Citympg (miles per gallon): là đơn vị đo lường hiệu quả nhiên liệu để xác định số dặm mà một xe ô tô có thể đi được trên một gallon nhiên liệu trong điều kiện lái trong thành phố.
+ Highwaympg: là một đơn vị đo lường hiệu quả nhiên liệu dùng để xác định số dặm mà một xe ô tô có thể đi được trên một gallon nhiên liệu trong điều kiện lái trên xa lộ hoặc đường cao tốc.
+ Price: giá của xe

# Xây dựng mô hình với toàn bộ data

```python
#Chọn dữ liệu X và y
y=df['price']
X=df.drop(columns = 'price', inplace = False)
```

```python
#Chia dữ liệu thành training set và test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# 1. Linear Regression
```

```python
#Xây dựng mô hình
from sklearn.linear_model import LinearRegression
model_lin= LinearRegression()
model_lin.fit(X_train,y_train)
print('a=',model_lin.coef_)
print('b=',model_lin.intercept_)
```

```python
#Dự báo với test set
y_pred_lin=model_lin.predict(X_test)
```

```python
#Đánh giá mô hình bằng chỉ số r-square
from sklearn.metrics import r2_score
Ac2_lin=r2_score(y_test, y_pred_lin)
print('R-square=',"{:.2f}".format(Ac2_lin))
```

```python
#Đánh giá mô hình bằng chỉ số root mean squared error(RMSE)
from sklearn.metrics import mean_squared_error
import math
rmse_lin = math.sqrt(mean_squared_error(y_test, y_pred_lin))
print('RMSE=',"{:.2f}".format(rmse_lin))
```
